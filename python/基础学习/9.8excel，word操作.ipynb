```python
from itertools import combinations
from functools import reduce
from openpyxl import load_workbook
#查找关系最好（共同参演电影数最多）的两、三个演员
def getdata(filename):
    actors=dict()
    ws=load_workbook(filename).worksheets[0]
    for index,row in enumerate(ws.rows):
        if index ==0:
            continue
        filename,actors=row[0].value,row[2].value.split(',')
        for a in actor:
            actors[a]=actors.get(a,set())
            actors[a].add(filename)
    return actors

def relations2(data,num):
    allfilms=reduce(lambda x:y:x|y,data.values(),set())
    combidata=combinations(data.items(),num)
    turelove=max(combidata,key=lambda item:len(reduce(lambda x,y:x&y,[i[1] for i in item],allfilms)))
    return('关系最好的{0}个演员是{1},','共同主演电影数量是{2}'.format(num,tuple(item[0] for item in turelove)),len(reduce(lambda x,y:x&y,[i[1] for it in turelove],allfilms)))
data=getdata('电影导演演员.xlsx')
print(relations2(data,2))
print(relations2(data,3))
```

```python
#合并单元格内容
from os import listdir
from os.path import exists
import openpyxl

result='result.xlsx'
if exists(result):
    os.remove(result)

wb_result=openpyxl.Workbook()
ws_result=wb_result.worksheet[0]
ws_result.append(['学院'，'姓名','成绩'])

for fn in (fns in fns in listdir() if fns.endwith('.xlsx')):
    wb=openpyxl.load_workbook(fn)
    ws=wb.worksheet[0]
    for index,row in enumerate(ws.rows):
        if index==0:
            continue
        ws_result.append(list(map(lambda cell:cell.value,row)))

rows=['']+list(ws_result.rows)
index1=2
rowcount=len(rows)

while index1<rowcount:
    value=rows[index1][0].value
    for index2,row2 in enumerate(rows[index1+1:],index1+1):
        if  not (row2[0].value==None or row2[0].value==value):
            break
        else:
            ws_result.merge_cells('A'+str(index1)+':A'+str(index2))
            break
    ws_result.merge_cells('A'+str(index1)+':A'+str(index2-1))
    index1=index2

wb_result.save(result)
```

```python
#txt转Excel文件
from os import listdir
import openpyxl

txts=[fn for fn in listdir() if fn.endswith('.txt')]
for fn in txts:
    newfn=fn[:-4]+'.xlsx'
    wb=openpyxl.Workbook()
    ws=wb.worksheets[0]
    with open(fn,encoding='cp936') as fp:
        for line in fp:
            line=line.strip().split(',')
            ws.append(line)
    wb.save(newfn)
```

```python
#批量设置Excel页眉页脚
import os
import openpyxl
from openpyxl.worksheet.header_footer import _HeaderFooterPart

xlsxfiles=(fn for fn in os.listdir('.') if fn.endswith('.xlsx'))
for xlsxfile in xlsxfiles:
    wb=openpyxl.load_workbook(xlsxfile)
    for ws in wb.worksheets:
        ws._HeaderFooter.differentFirst=True
        ws._HeaderFooter.differentOddEven=True
        ws.firstHeader.left=_HeaderFooterPart('第一页左页眉'，size=24,color='FF0000')
        ws.firstFooter.center=_HeaderFooterPart('第一页中页眉'，size=24,color='00FF00')
        ws.oddHeader.right=_HeaderFooterPart('奇数页右页眉')
        ws.oddFooter.center=_HeaderFooterPart('奇数页中页眉')

        ws.evenHeader.left=_HeaderFooterPart('偶数页左页眉')
        ws.evenFooter.center=_HeaderFooterPart('偶数页中页眉')
wb.save('new_'+xlsxfile)
```

```python
#设置字体，背景
from random import sample
import openpyxl
from openpyxl.styles import Font,colors,Fill

def generateXlsx(num):
    for i in range(num):
        wb=openpyxl.Workbook()
        ws=wb.worksheets[0]
        ws.append(['字段'+str(_) for _ in range(1,6)])
        for _ in range(10):
            ws.append(sample(range(10000),5))
        wb.save(str(i)+'.xlsx')

def batch_format(num):
    for i in range(num):
        fn=str(i)+'.xlsx'
        wb=openpyxl.load_workbook(fn)
        ws=wb.worksheets[0]
        for irow,row in enumerate(ws.rows,start=1):
            if irow==1:
                font=Font('黑体'，bold=True)
            elif irow%2==0:
                font=Font('宋体'，color=colors.RED)
            else:
                font=Font('宋体'，color='00CCFF')
            for cell in row:
                cell.font=font
                if irow%2==0:
                    cell.fill=openpyxl.style.fills.GradientFill(stop=['FF0000','0000FF'])
    wb.save('new'+fn)

generateXlsx(5)
batch_format(5)
```

```python
#添加成绩到Excel文件
from random import choice,randint
from openpyxl import Workbook,load_workbook

def generate_random_information(filename):
    workbook = Workbook()
    worksheet = workbook.worksheets[0]
    worksheet.append(['姓名','课程','成绩'])

    first='李刘张孙钱'
    middle='休年假瞌睡虫'
    last='今年初的借'
    subjects=('语文'，'数学','英语')

    for i in range(20):
        name=choice(first)
        if randint(1,100)>50:
            name=name+choice(middle)
        name=name+choice(last)

        worksheet.append([name,choice(subjects),randint(0,100)])
    worksheet.save(filename)

def getresults(oldfile,newfile):
    results=dict()
    worksheet=load_workbook(oldfile).worksheet[0]

    for row in worksheet.rows:
        if row[0].value=='姓名':
            continue
        name,subjects,grade=map(lambda cell:cell.value,row)
        t=results.get(name,{})
        f=t.get(subjects,0)

        t[subjects]=max(grade,f)
        results[name]=t

    workbook1=Workbook()
    worksheet1=workbook1.worksheets[0]
    worksheet1.append(['姓名','课程','成绩'])

    for name,t in results.items():
        for subjects,grade in t.items():
            worksheet1.append([name,subjects,grade])
    workbook1.save(newfile)

oldfile=r''
newfile=r''
generate_random_information(oldfile)
getresults(oldfile,newfile)


```

```python
#提取word文档字，图片，表格
from docx import Document
import re

doc=Document('text.docx')
result={'li':[],'fig':[],'tab':[]}

for p in doc.paragraphs:
    t=p.text
    if re.match('例\d+-\d+',t):
        result['li'].append(t)
    elif re.match('图\d+-\d+',t):
        result['fig'].append(t)
     elif re.match('表\d+-\d+',t):
        result['tab'].append(t)   
for key in result.keys():
    print('='*30)
    for value in result[key]:
        print(value)
```
